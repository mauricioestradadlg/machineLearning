**Fusi贸n de Datos: Crea un segundo DataFrame que contenga los datos del archivo Datos_Ventas_Tienda2.csv (tambi茅n provisto en esta lecci贸n), y concat茅nalos para tener un solo dataFrame con toda la informaci贸n.**

```python
df = pd.concat([df1, df2])
df
```

**Tratamiento de Datos: Utiliza Pandas para manipular estos datos. Deber谩s realizar tareas como limpieza de datos, filtrado y transformaciones b谩sicas.**

```python
df['Fecha'] = pd.to_datetime(df['Fecha'], format='%m/%d/%Y')
```

```python
df_agrupado = df.groupby('Producto')['Total Venta'].mean()
df_agrupado
```

**An谩lisis de Ventas: Realiza an谩lisis para responder preguntas como:**


1. 驴Cu谩l es el producto m谩s vendido?

2. 驴Cu谩l es el mes con m谩s ventas?

```python
# Paso 1: Obtener el valor m谩ximo
valor_max = df['Total Venta'].max()

# Paso 2: Filas que tienen ese valor m谩ximo
filas_max = df[df['Total Venta'] == valor_max]

# Paso 3: Imprimir detalles claros
print("\n Producto(s) con la venta m谩s alta del DataFrame:")
for idx, fila in filas_max.iterrows():
    print(f"ndice: {idx}, Producto: {fila['Producto']}, Total Venta: {fila['Total Venta']}")
```

```python
# Paso 1: Asegurarnos de que la columna 'Fecha' sea de tipo datetime
df['Fecha'] = pd.to_datetime(df['Fecha'])

# Paso 2: Crear una nueva columna con el mes en formato YYYY-MM
df['Mes'] = df['Fecha'].dt.to_period('M').astype(str)

# Paso 3: Agrupar por mes y calcular la suma de "Total Venta"
ventas_por_mes = df.groupby('Mes')['Total Venta'].sum().reset_index()

# Paso 4: Obtener el valor m谩ximo y las filas correspondientes
valor_max = ventas_por_mes['Total Venta'].max()
meses_max = ventas_por_mes[ventas_por_mes['Total Venta'] == valor_max]

# Paso 5: Imprimir resultados legibles
print("\n Mes(es) con la mayor suma de ventas:")
for _, fila in meses_max.iterrows():
    print(f"Mes: {fila['Mes']}, Total Venta: {fila['Total Venta']}")
```
